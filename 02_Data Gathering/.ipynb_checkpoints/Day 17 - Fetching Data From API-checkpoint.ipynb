```python
import requests
import pandas as pd
import time

url = "https://real-time-amazon-data.p.rapidapi.com/product-reviews"

headers = {
    "x-rapidapi-key": "507b37df13msh479715189576570p16bfd8jsndb35155ca1b2",
    "x-rapidapi-host": "real-time-amazon-data.p.rapidapi.com"
}

# Start with empty DataFrame with required columns
df = pd.DataFrame(columns=["review_id", "review_title", "review_star_rating", "review_author_id", "review_author"])

# Loop through pages (adjust max pages as needed)
for i in range(1, 51):  
    querystring = {
        "asin": "B07ZPKN6YR",
        "country": "US",
        "page": str(i),
        "sort_by": "TOP_REVIEWS",
        "star_rating": "ALL",
        "verified_purchases_only": "false",
        "images_or_videos_only": "false",
        "current_format_only": "false"
    }

    response = requests.get(url, headers=headers, params=querystring)

    if response.status_code == 200:
        reviews = response.json().get("data", {}).get("reviews", [])
        if reviews:
            temp_df = pd.DataFrame(reviews)[
                ["review_id", "review_title", "review_star_rating", "review_author_id", "review_author"]
            ]
            df = pd.concat([df, temp_df], ignore_index=True)
        else:
            print(f"No more reviews on page {i}, stopping loop.")
            break
    else:
        print(f"Request failed on page {i}, status code:", response.status_code)
        break

    time.sleep(0.2)  # avoid hitting API rate limit

print(df.head())
print("Total reviews collected:", len(df))
```

```python
import requests
import pandas as pd

url = "https://real-time-amazon-data.p.rapidapi.com/product-reviews"
headers = {
    "x-rapidapi-key": "507b37df13msh479715189576570p16bfd8jsndb35155ca1b2",
    "x-rapidapi-host": "real-time-amazon-data.p.rapidapi.com"
}

# Collect reviews here
all_reviews = []

# Loop only first 10 pages (API hard limit)
for i in range(1, 21):
    params = {
        "asin": "B07ZPKN6YR",
        "country": "US",
        "page": str(i),
        "sort_by": "TOP_REVIEWS",
        "star_rating": "ALL",
        "verified_purchases_only": "false",
        "images_or_videos_only": "false",
        "current_format_only": "false"
    }

    r = requests.get(url, headers=headers, params=params)
    if r.status_code != 200:
        print(f"Error on page {i}: {r.status_code}")
        break

    reviews = r.json().get("data", {}).get("reviews", [])
    if not reviews:  # stop if empty
        break

    all_reviews.extend(reviews)

# Convert to DataFrame with only required columns
df = pd.DataFrame(all_reviews)[
    ["review_id", "review_title", "review_star_rating", "review_author_id", "review_author"]
]

print(df.head())
print("Total reviews collected:", len(df))
```
