```python
import requests

url = "https://imdb8.p.rapidapi.com/title/v2/get-awards"

querystring = {"tconst":"tt0120338","first":"20","country":"US","language":"en-US"}

headers = {
	"x-rapidapi-key": "507b37df13msh479715189576570p16bfd8jsndb35155ca1b2",
	"x-rapidapi-host": "imdb8.p.rapidapi.com"
}

response = requests.get(url, headers=headers, params=querystring)

print(response.json())
```
