```python
import pandas as pd
import seaborn as sns
```

```python
tips = pd.read_csv('tip.csv')
tips.head()
```

```python
df = pd.read_csv('Titanic-Dataset.csv')
```

## 1] Scatterplot(Numerical - Numerical)

```python
sns.scatterplot(x='total_bill',y='tip',data=tips)
```

```python
sns.scatterplot(x=tips['total_bill'],y=tips['tip'])
```

```python
sns.scatterplot(x='total_bill',y='tip',data=tips,hue='sex',style='smoker',size='size')
```

## Barplot(Categorical - Numerical)

```python
sns.barplot(x='Pclass',y='Fare',data=df,hue='Embarked')
```

## Boxplot

```python
sns.boxplot(x='Sex',y='Age',data=df,hue='Embarked')
```

## Distplot

```python
sns.distplot(df[df['Survived']==0]['Age'],hist=False)
sns.distplot(df[df['Survived']==1]['Age'],hist=False)
```

## Heatmap

```python
sns.heatmap(pd.crosstab(df['Pclass'],df['Survived']))
```

```python
(df.groupby('Pclass')['Survived'].mean()*100).plot(kind='bar')
```

## Clustermap

```python
sns.clustermap(pd.crosstab(df['Parch'],df['Survived']))
```
