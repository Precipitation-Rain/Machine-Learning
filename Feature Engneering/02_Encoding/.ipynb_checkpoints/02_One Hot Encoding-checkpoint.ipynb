### Independent and dependent varaible

```python
y = df.iloc[:,-1]
```

```python
x
```

### train_test_split

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2)
```

```python
x_train.shape
```

```python
x_test
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
ohe = OneHotEncoder(drop='first') # drop = 'first' is used to deal with problem of multicolinearity
#sparse_output = False will no return a sparse matrix so you don't need to do toaaray()
#dtype = np.int32 -> makes all float to int
#you may need all parameters that i commented out
```

```python
ohe.fit(x_train[['fuel','owner']])
```

```python
x_train_ohe = ohe.transform(x_train[['fuel','owner']]).toarray()
```

```python
x_test_ohe = ohe.transform(x_test[['fuel','owner']]).toarray()
```
