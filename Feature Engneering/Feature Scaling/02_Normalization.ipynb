```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('wine_data.csv',header=None,usecols=[0,1,2])
df.columns=['Class Label','Alchole','Malic acid']
df.sample(5)
```

```python
sns.kdeplot(data=df,x='Alchole')
plt.show()
```

```python
sns.kdeplot(data=df,x='Malic acid')
plt.show()
```

```python
sns.scatterplot(x='Alchole',y='Malic acid',hue='Class Label',data=df)
plt.show()
```

```python
x = df.drop('Class Label',axis=1)
y = df['Class Label']
```

## train_test_split

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=0)# follo x_train,x_test,y_train,y_test sequence
```

```python
x_train.shape,x_test.shape
```

## Model Training

```python
from sklearn.preprocessing import MinMaxScaler
mms = MinMaxScaler()

mms.fit(x_train)

x_train_scaled = mms.transform(x_train)
x_test_scaled = mms.transform(x_test)
```

```python
x_train_scaled = pd.DataFrame(x_train_scaled,columns=x_train.columns)
x_test_scaled = pd.DataFrame(x_test_scaled,columns=x_test.columns)
```

```python
np.round(x_train.describe(),1)
```

```python
np.round(x_train_scaled.describe(),1)
```

```python
np.round(x_test_scaled.describe(),1)
```

```python
fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(12, 5))

ax1.scatter(x_train['Alchole'], x_train['Malic acid'], c=y_train)
ax1.set_title("Before Scaling")
ax2.scatter(x_train_scaled['Alchole'], x_train_scaled['Malic acid'], c=y_train)
ax2.set_title("After Scaling")
plt.show()
```

```python
fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(12, 5))

# before scaling
ax1.set_title('Before Scaling')
sns.kdeplot(x_train['Alchole'], ax=ax1)
sns.kdeplot(x_train['Malic acid'], ax=ax1)

# after scaling
ax2.set_title('After Standard Scaling')
sns.kdeplot(x_train_scaled['Alchole'], ax=ax2)
sns.kdeplot(x_train_scaled['Malic acid'], ax=ax2)
plt.show()
```

```python
fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(12, 5))

# before scaling
ax1.set_title('Alchole Distribution Before Scaling')
sns.kdeplot(x_train['Alchole'], ax=ax1)

# after scaling
ax2.set_title('Alchole Distribution After Standard Scaling')
sns.kdeplot(x_train_scaled['Alchole'], ax=ax2)
plt.show()
```

```python
fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(12, 5))

# before scaling
ax1.set_title('Malic acid Distribution Before Scaling')
sns.kdeplot(x_train['Malic acid'], ax=ax1)

# after scaling
ax2.set_title('Malic acid Distribution After Standard Scaling')
sns.kdeplot(x_train_scaled['Malic acid'], ax=ax2)
plt.show()
```
