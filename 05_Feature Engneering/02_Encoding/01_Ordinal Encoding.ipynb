```python
import pandas as pd
```

```python
df = pd.read_csv('customer.csv')
df.head()
```

```python
x= df.iloc[:,2:4]
y = df['purchased']
```

```python
# It is must that and highly recommended that you should split data into train and test before doing feature transformation
```

## Train_test_split

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
x_train.shape,x_test.shape
```

```python
x_train
```

```python
y_train
```

## Ordinal Encoding

```python
# used for ordinal categorical data(have some order between data. ex] pg > ug > hs) only for input columns
```

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
oe = OrdinalEncoder(categories=[['Poor','Average','Good'],['School','UG','PG']])
```

```python
oe.fit(x_train)
```

```python
oe.transform(x_train)
oe.transform(x_test)
```

## Label Encoding

```python
# used for ordinal categorical data(have some order between data. ex] pg > ug > hs) only for output columns
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
```

```python
le.fit(y_train)
```

```python
le.transform(y_train)
```

```python
le.transform(y_test)
```
